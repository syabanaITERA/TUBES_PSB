```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

# Fungsi untuk membuat sinyal x1(t) dan x2(t)
def x1(t):
    return 3 * np.cos(2 * np.pi * 10 * t - 0.2 * np.pi)

def x2(t):
    return np.sin(2 * np.pi * 5 * t)

# Membuat waktu (t) dari 0 hingga 2 detik dengan langkah 0.001 detik
t = np.arange(0, 2, 0.001)

# Membuat sinyal x1(t) dan x2(t)
signal1 = x1(t)
signal2 = x2(t)

# Menggabungkan kedua sinyal
combined_signal = signal1 + signal2

# Menghitung spektrogram
frequencies, times, Sxx = spectrogram(combined_signal, fs=1 / 0.001)

# Plot sinyal x1(t), x2(t), dan gabungan
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal1)
plt.title('Sinyal x1(t)')

plt.subplot(3, 1, 2)
plt.plot(t, signal2)
plt.title('Sinyal x2(t)')

plt.subplot(3, 1, 3)
plt.plot(t, combined_signal)
plt.title('Gabungan Sinyal x1(t) dan x2(t)')

plt.tight_layout()
plt.show()

# Plot spektrogram
plt.figure(figsize=(10, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='auto')
plt.colorbar(label='Intensity (dB)')
plt.title('Spektrogram dari Gabungan Sinyal')
plt.xlabel('Waktu (detik)')
plt.ylabel('Frekuensi (Hz)')
plt.show()

# Plot domain frekuensi
plt.figure(figsize=(10, 6))
plt.magnitude_spectrum(combined_signal, Fs=1 / 0.001, scale='dB', color='blue')
plt.title('Domain Frekuensi dari Gabungan Sinyal')
plt.xlabel('Frekuensi (Hz)')
plt.ylabel('Amplitudo (dB)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat sinyal x1(t) dan x2(t)
def x1(t):
    return 3 * np.cos(2 * np.pi * 10 * t - 0.2 * np.pi)

def x2(t):
    return np.sin(2 * np.pi * 5 * t)

# Membuat waktu (t) dari 0 hingga 2 detik dengan langkah 0.001 detik
t = np.arange(0, 2, 0.001)

# Membuat sinyal x1(t) dan x2(t)
signal1 = x1(t)
signal2 = x2(t)

# Menggabungkan kedua sinyal
combined_signal = signal1 + signal2

# Transformasi Fourier untuk mendapatkan domain frekuensi
frequencies = np.fft.fftfreq(len(t), 0.001)  # Frekuensi dalam Hz
fft_values = np.fft.fft(combined_signal)

# Plot domain frekuensi
plt.figure(figsize=(10, 6))
plt.stem(frequencies, np.abs(fft_values))
plt.title('Domain Frekuensi dari Gabungan Sinyal')
plt.xlim(-20, 20)
plt.xlabel('Frekuensi (Hz)')
plt.ylabel('Amplitudo')
plt.grid(True)
plt.show()
```
